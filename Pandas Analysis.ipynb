```python
# Adding a new column for 'Total' stats
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
# Removing columns from the data
df = df.drop(columns=['Total'])
print(df)
```

```python
# Adding column, a different way
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
print(df)
```

```python
df.head(5)
```

```python
# Re ordering table 
cols = list(df.columns)
df = df[cols[0:10] + [cols[-1]] + cols[10:12]]
df.head(5)
# Can also be done by naming all the column headings in a list like so
# df = df[['Total', 'HP', 'Name' etc.. ]]
```

```python
# Filtering Data
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
# Saving the filtered data as a new Data Frame within Python, and exporting as a File type
Filtered_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') | (df['Total'] > 500)]
#Filtered_df.to_csv('Filtered_Poke.csv', index=False)
print(Filtered_df)
```

```python
# Reset Index on the newly filtered Data Frame
Filtered_df.reset_index(drop=True, inplace=True)
print(Filtered_df)
```

```python
# Filtering things out of the Data Frame
import re
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# Further filtering
# df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I,regex=True)]
df.loc[df['Name'].str.contains('Pi[a-z*]')]
```

```python
# Conditional Formatting
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flame Boys' 
df.loc[df['Type 2'] == 'Fire', 'Type 2'] = 'Flame Boys' 
df.loc[df['Type 1'] == 'Grass', 'Type 1'] = 'Flower Boys'
df.loc[df['Type 2'] == 'Grass', 'Type 2'] = 'Flower Boys' 
df.loc[df['HP'] == '', 'Type 1'] = 'Flame Boys' 
df.loc[(df['Type 2'] == 'Flower Boys') | (df['Type 1'] == 'Flame Boys')]
```
